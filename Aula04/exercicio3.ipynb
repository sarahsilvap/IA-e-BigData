```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
A = np.array([50, 100, 120, 300]).reshape(-1, 1)  # Área dos imóveis em metros quadrados
Valor = np.array([180000, 300000, 375000, 600000])  # Valor dos imóveis em reais
```

```python
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(A, Valor)
```

```python
# Dados para previsão
A_teste = np.array([70, 150, 250, 350]).reshape(-1, 1)  # Áreas para previsão
```

```python
# Fazer previsões com base na área
Valor_previsto = modelo.predict(A_teste)
```

```python
# Imprimir as previsões
for i in range(len(A_teste)):
    print(f"Área: {A_teste[i][0]} m² - Valor previsto: R${Valor_previsto[i]:,.2f}")

# Plotar os dados
plt.scatter(A, Valor, color='blue', label='Dados de Treinamento')
plt.plot(A, modelo.predict(A), color='green', label='Linha de Regressão')
plt.scatter(A_teste, Valor_previsto, color='red', label='Dados de Teste')
plt.xlabel('Área (m²)')
plt.ylabel('Valor (R$)')
plt.title('Relação entre Área e Valor do Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
