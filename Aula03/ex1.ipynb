```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0, 10, 10)
dados_y = dados_x * dados_x
```

```python
plt.figure(figsize=(10, 5), facecolor=('grey'))
Ax = plt.axes() #cria os eixos do 
Ax.set_facecolor('white')
plt.plot(dados_x, dados_y, color = 'red', linewidth = 2, marker = 'o', linestyle = 'dashed', markeredgecolor = 'black')
plt.xlabel('Eixo x', fontsize = 12)
plt.ylabel('Eixo y', fontsize = 12)
plt.title('Exemplo Matplotlib', fontsize = 12)
```
