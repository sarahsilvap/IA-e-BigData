```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir os nomes das lojas e os dados de vendas
lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criar o DataFrame a partir dos dados de vendas
df_vendas = pd.DataFrame(vendas_2022, index=lojas)

# Criar a figura com subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 linhas, 2 colunas de subplots

# Ajustar o layout para evitar sobreposio
plt.tight_layout(pad=5.0)

# Iterar sobre cada loja e subplot
for i, loja in enumerate(lojas):
    ax = axs[i // 2, i % 2]  # Selecionar o subplot correto
    ax.plot(df_vendas.columns, df_vendas.loc[loja], marker='o', linestyle='-', label=f'Loja {loja}')
    ax.set_title(f'Vendas na Loja {loja}', fontsize=14)
    ax.set_xlabel('Meses', fontsize=12)
    ax.set_ylabel('Nmero de Vendas', fontsize=12)
    ax.legend()
    ax.grid(True)

# Ajustar o ttulo da figura
fig.suptitle('Variao no Nmero de Vendas nas Lojas ao Longo de 2022', fontsize=16)

# Exibir a figura
plt.show()
```
