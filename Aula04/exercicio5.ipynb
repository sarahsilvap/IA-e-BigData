```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)  # Área dos imóveis em metros quadrados
demanda = np.array([100, 200, 80, 100, 160])  # Valor dos imóveis em reais
```

```python
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(consumo, demanda)
```

```python
# Dados para previsão
teste = np.array([35, 45, 150, 60, 75]).reshape(-1, 1)  # Áreas para previsão
```

```python
# Fazer previsões 
demanda_prevista = modelo.predict(teste)
```

```python
# Imprimir as previsões
for i in range(len(teste)):
    print(f"Área: {teste[i][0]} m² - Valor previsto: R${demanda_prevista[i]:,.2f}")

# Plotar os dados
plt.scatter(consumo, demanda, color='blue', label='Dados de Treinamento')
plt.plot(consumo, modelo.predict(consumo), color='green', label='Linha de Regressão')
plt.scatter(teste, demanda_prevista, color='red', label='Dados de Teste')
plt.xlabel('Consumo (KW)')
plt.ylabel('Demanda')
plt.title('Relação entre Consumo (em KW) e Demanda de energia')
plt.legend()
plt.grid(True)
plt.show()
```
