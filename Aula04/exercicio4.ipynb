```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)  # Área dos imóveis em metros quadrados
gastos = np.array([750, 1500, 4000, 6000, 18000])  # Valor dos imóveis em reais
```

```python
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(renda, gastos)
```

```python
# Dados para previsão
teste = np.array([1000, 7500, 15000, 28000]).reshape(-1, 1)  # Áreas para previsão
```

```python
# Fazer previsões 
gastos_previsto = modelo.predict(teste)
```

```python
# Imprimir as previsões
for i in range(len(teste)):
    print(f"Área: {teste[i][0]} m² - Valor previsto: R${gastos_previsto[i]:,.2f}")

# Plotar os dados
plt.scatter(renda, gastos, color='blue', label='Dados de Treinamento')
plt.plot(renda, modelo.predict(renda), color='green', label='Linha de Regressão')
plt.scatter(teste, gastos_previsto, color='red', label='Dados de Teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação entre Renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```
