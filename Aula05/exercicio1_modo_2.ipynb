```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Definir os nomes das lojas
lojas = np.array(['A', 'B', 'C', 'D'])

# Definir os dados de vendas para 2022 como um dicion√°rio
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 190, 350, 200],
    'Mar': [220, 190, 350, 200],
    'Abr': [230, 150, 280, 120],
    'Mai': [250, 170, 300, 140],
    'Jun': [260, 180, 310, 150],
    'Jul': [240, 160, 290, 130],
    'Ago': [220, 140, 270, 110],
    'Set': [400, 220, 350, 190],
    'Out': [300, 350, 400, 250]
}
```

```python
# Criar um DataFrame a partir dos dados de vendas
df_vendas = pd.DataFrame(vendas_2022, index=lojas)

# Exibir o DataFrame
print(df_vendas)
```

```python
# Transpor o DataFrame para que os meses sejam as linhas e as lojas as colunas
df_vendas = df_vendas.T

# Exibir o DataFrame para verificar sua estrutura
print(df_vendas)

plt.figure(figsize=(12, 6))
```

```python
# Plotar cada loja
for loja in df_vendas.columns:
    plt.plot(df_vendas.index, df_vendas[loja], label=f'Loja {loja}')

# Configurar o gr√°fico
plt.xticks(rotation=45, fontsize=12)  # Rotacionar os r√≥tulos dos meses para melhor visualiza√ß√£o e ajustar o tamanho da fonte
plt.yticks(fontsize=12)  # Ajustar o tamanho da fonte para os valores do eixo y
plt.xlabel('Meses', fontsize=14)  # T√≠tulo do eixo x
plt.ylabel('N√∫mero de Vendas', fontsize=14)  # T√≠tulo do eixo y
plt.title('Vendas Mensais das Lojas em 2022', fontsize=18)  # T√≠tulo do gr√°fico
plt.legend(title='Lojas')
plt.grid(True)

# Exibir o gr√°fico
plt.show()
```
