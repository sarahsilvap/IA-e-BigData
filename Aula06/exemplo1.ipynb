```python
import numpy as np
from sklearn.model_selection import train_test_split #importa da sklearn.model a bibliotexa train test split 
#essa biblioteca serve para dividir os dados teste e treino
from sklearn.tree import DecisionTreeClassifier #vai importar o algorítimo da árvore de decisão
from sklearn import tree #importa biblioteca para plotar a árvore
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Dados treinamento
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo','pessimo',
                    'bom','pessimo', 'pessimo','ruim','ruim','bom','bom'])
```

```python
#dividir dados em teste e treino
notas_treino, notas_teste, classes_treino, classes_teste = train_test_split(notas.reshape(-1,1), classes.reshape(-1,1), test_size=0.2, random_state=42)           
```

```python
modelo = DecisionTreeClassifier() #crio o modelo de árvore de decisão
modelo.fit(notas_treino,classes_treino) #treinamento do modelo de árvore de decisão
```

```python
previsoes = modelo.predict(notas_teste) #dentro de nodelo predict é colocado as notas de teste
previsoes
```

```python
notas_teste
```

```python
#teste de nota individual
nota = np.array([4, 7]).reshape(-1,1)
resultado = modelo.predict(nota)
resultado
```

```python
fig = plt.figure(figsize=(10,8)) #cria uma figura de tamanho 10 x 8
#tree plot plota a árvode de decisão
#modelo é um modelo criado do algoritmo de aprendizado supevisado
tree.plot_tree(modelo,feature_names= notas.tolist(), class_names = classes.tolist(), filled= True)
plt.title('Árvore de decisão para classificar uma bebida \n de acordo com a nota', fontsize=18)
plt.show()
```

```python
#planta box plot com a bibliotexa seabrond
# x é o eixo x que indica as classes: boa, ruim ou péssimo
# y indica  as notas

plt.figure(figsize=(10,8))
#plt.grid()
plt.title('Box plot das notas e classes das bebidas')
plt.ylabel('Notas')
plt.xlabel('Classes')

sns.boxplot(x = classes, y = notas)
```

```python
#frafico de barras
plt.figure(figsize=(10,8))
sns.barplot(y = notas, x = classes)
plt.title('Notas de bebidas', fontsize = 20)
plt.xlabel('Classicação', fontsize = 16)
plt.ylabel('Notas', fontsize = 16)
```
